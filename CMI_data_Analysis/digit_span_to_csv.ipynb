```python
n_subs=40
betas_avg_EC_mf=[]
betas_avg_EO_mf=[]
betas_avg_EC_po=[]
betas_avg_EO_po=[]
for i in range(n_subs):
    this_sub_row=new.iloc[i]
    
    betas_mf_EC=[this_sub_row[7],this_sub_row[11],this_sub_row[15],this_sub_row[19]]
    betas_avg_EC_mf.append(betas_mf_EC)
    betas_mf_EO=[this_sub_row[9],this_sub_row[13],this_sub_row[17],this_sub_row[21]]
    betas_avg_EO_mf.append(betas_mf_EO)
    
    betas_po_EC=[this_sub_row[8],this_sub_row[12],this_sub_row[16],this_sub_row[20]]
    betas_avg_EC_po.append(betas_po_EC)
    betas_po_EO=[this_sub_row[10],this_sub_row[14],this_sub_row[18],this_sub_row[22]]
    betas_avg_EO_po.append(betas_po_EO)
```

```python
new['betas_avg_EC_mf']=betas_avg_EC_mf
```

```python
new
```

```python
import numpy as np
```

```python
collapse_EC_mf=[np.mean(i) for i in betas_avg_EC_mf]
collapse_EC_po=[np.mean(i) for i in betas_avg_EC_po]
collapse_EO_mf=[np.mean(i) for i in betas_avg_EO_mf]
collapse_EO_po=[np.mean(i) for i in betas_avg_EO_po]
```

```python
collapse_EC_mf[0]==np.mean(betas_avg_EC_mf[0])
```

```python
new['betas_avg_EC_mf']=collapse_EC_mf
new['betas_avg_EC_po']=collapse_EC_po
new['betas_avg_EO_mf']=collapse_EO_mf
new['betas_avg_EO_po']=collapse_EO_po
```

```python
tb_EC_mf=[]
tb_EO_mf=[]
tb_EC_po=[]
tb_EO_po=[]
for sub in range(40):
    subz_row=new.iloc[sub]
    tb_EC_mf.append(subz_row['THETA_psds_EC_mf0']/subz_row['betas_avg_EC_mf'])
    tb_EO_mf.append(subz_row['THETA_psds_EO_mf0']/subz_row['betas_avg_EO_mf'])
    tb_EC_po.append(subz_row['THETA_psds_EC_po0']/subz_row['betas_avg_EC_po'])
    tb_EO_po.append(subz_row['THETA_psds_EO_po0']/subz_row['betas_avg_EO_po'])
```

```python
new['THETA_OVER_BETA_EC_mf']=tb_EC_mf
new['THETA_OVER_BETA_EO_mf']=tb_EO_mf
new['THETA_OVER_BETA_EC_po']=tb_EC_po
new['THETA_OVER_BETA_EO_po']=tb_EO_po
```
