```python
r4subs_to_keep=list(r4_no_diags)[:]
del_count=0
for n in range(1,len(r4_no_diags)):
    subject=r4_no_diags[n]
    allS_ind=list(all_IDs).index(subject)+1
    age=float(all_ages[allS_ind])
    completion=all_drop_outs[allS_ind]
    print(allS_ind)
    print(age)
    print(subject)
    #print(completion)
    if 25>age>5:
        right_age=1
    else:
        right_age=0
    #if no_diags ==0:
        #print('\n\n')
        #print(age)
        #print(completion)
        #print(sub_diag)
    if (subject not in noDXcon) or right_age==0 or (completion != 'Complete'):
        r4_subs_to_keep.remove(subject)
        del_count=del_count+1
```
