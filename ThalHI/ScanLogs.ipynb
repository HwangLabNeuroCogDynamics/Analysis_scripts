## Converting behavioral logs of MRI data to '.1D' text files for use with AFNI

```python
import numpy as np  # whole numpy lib is available, prepend 'np.'
from numpy import average, std
from numpy.random import random, randint, normal, shuffle,uniform
import seaborn as sns
import fnmatch
import os  # handy system and path functions
import sys  # to get file system encoding
import csv
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd  
import matplotlib
```

```python
sub='110'
```

```python
ROOT='/home/dcellier/RDSS/ThalHi_data/MRI_data/behavioral_data/'
filen_pattern='%s_00*_Task_THHS_20*_*_*.csv'
```

```python
thisSub_runs={}
for behav_file in os.listdir(ROOT):
    if fnmatch.fnmatch(behav_file,filen_pattern %sub):
        print(behav_file)
        run_num=behav_file.split('_')[1]
        thisSub_runs[run_num]=pd.read_csv(ROOT+behav_file,engine='python')
    
```

## Compiling stay trial info for ea block

```python
stay=pd.DataFrame()

for n in range(1,6):
    block='00'+str(n)
    stay=stay.append(thisSub_runs[block][thisSub_runs[block].Trial_type=='Stay'].Time_Since_Run_Cue_Prez.reset_index(drop=True))
```
