```python
row_info={}
for row in range(len(behav_files)):
    if behav_files.iloc[row]['trial_n']<10: #isolating the first 10 trials of each block because it's only the beginning of ea block that was problematic
        trig_dict=ast.literal_eval(behav_files['trigs'][row]) # getting the trigger events from each row
        if 'Stay_trig' in trig_dict:
            c=trig_dict['Stay_trig']
        elif 'IDS_trig' in trig_dict:
            c=trig_dict['IDS_trig']
        elif 'EDS_trig' in trig_dict:
            c=trig_dict['EDS_trig']
        if 'Face' in trig_dict:
            p=trig_dict['Face']
        elif 'Scene' in trig_dict:
            p=trig_dict['Scene']
        if 'resp' in trig_dict:
            r=trig_dict['resp']
        elif 'NoResp' in trig_dict:
            r=trig_dict['NoResp']
        row_num=row
        row_info[behav_files.iloc[row]['block']+'_'+str(behav_files.iloc[row]['trial_n'])]={'block':int(behav_files.iloc[row]['block']),'block_trial_n':behav_files.iloc[row]['trial_n'],'index':row_num,'cue':c,'probe':p,'resp':r} # saving out the trigggers and index of this row to be matched to triggers from bdf files
    else:
        continue
    
```

```python
skipped_rows=0
which_block=1
which_rows_to_delete={}
eeg_blocks=[block1,block2,block3,block4,block5]
for eeg_trigs in eeg_blocks: #looping through each block of trigs
    i=0 #index of first event in eeg_trigs
    e=eeg_trigs[i]
    while not (e==105 or e==103 or e==101): #if we're at a weird trigger, continue
        print(e)
        i=i+1
        e=eeg_trigs[i]
    not_found=1
    while not_found==1:
        if e==105 or e==103 or e==101: # if this event is a stay,IDS,orEDS trig
            nextE=ev_list[i+1]
            if nextE==153 or nextE==151: # and the next event is a probe
                nextnextE=ev_list[i+2]
                if nextnextE==155 or nextnextE==157: # and the next next event is a response
                    iti=ev_list[i+3]
                    if iti==159: # and the final event is an iti
                        print('\n')
                        print(e,nextE,nextnextE,iti)
                        for trial in range(10): # running through the keys of row_info, which only took the first 10 trials of ea block
                            key='00'+str(which_block)+'_'+str(trial)
                            this_trial_dict=row_info[key]
                            print(this_trial_dict)
                            cue=this_trial_dict['cue']
                            probe=this_trial_dict['probe']
                            response=this_trial_dict['resp']
                            if not (e==cue and nextE==probe and nextnextE==response): # if this trial is not documented by the eeg triggers, add to skipped_rows
                                skipped_rows=skipped_rows+1
                            elif (e==cue and nextE==probe and nextnextE==response): #else if this is the first trial recorded in eeg, save out the trial index
                                which_rows_to_delete[str(which_block)]={'lost_rows':skipped_rows,'first_ind':this_trial_dict['index']}
                                print(which_rows_to_delete[str(which_block)])
                                skipped_rows=0
                                which_block=which_block+1 # move onto the next block
                                #skip through eeg triggers to the next block
                                not_found=0
                                print('\n')

        else:
            print('error')
                
```
